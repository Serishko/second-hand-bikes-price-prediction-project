```python
import pandas as pd
import numpy as np
import os
```

## Listing all the files containing the names of bikes

```python
path=r'C:\Users\santo\Desktop\motorbikes'
bikeslist = os.listdir(path)
bikeslist
```

## Merging all the files into one csv file

```python
df=[]
for i in bikeslist:
    data = pd.read_csv(f"C:\\Users\\santo\\Desktop\\motorbikes\\{i}")
    data['Brand'] = i[:-4]
    a = ['Brand', 'Name', 'Lot No.', 'Kilometers', 'Price']
    data = data[a]
    df.append(data)
```

```python
final_data = pd.concat(df)
```

```python
final_data
```

## Dropping all the duplicate rows

```python
final_data.drop_duplicates(inplace=True)
final_data.reset_index(inplace=True)
final_data.drop(columns=['index'], inplace=True)
final_data
```

## Verifying if our data contains the column name

```python
final_data[final_data['Name']=='Name']
```

## Dropping down the indices with such column names

```python
final_data.drop(index=[495,2186,4343], inplace=True)
final_data.reset_index(drop=True, inplace=True)
final_data
```

```python
final_data[final_data['Name']=='Name']
```

## Saving the final data in csv

```python
final_data.to_csv(r'C:\Users\santo\PycharmProjects\MachineLearning\DataSets for ML\hamrobazar bikeslist\data.csv', index=False)
```
